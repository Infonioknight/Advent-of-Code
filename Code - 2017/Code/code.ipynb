```python
ins = {'A': {0: [1, 1, 'B'], 1: [0, -1, 'C']}, 
       'B': {0: [1, -1, 'A'], 1: [1, 1, 'D']}, 
       'C': {0: [1, 1, 'A'], 1: [0, -1, 'E']}, 
       'D': {0: [1, 1, 'A'], 1: [0, 1, 'B']}, 
       'E': {0: [1, -1, 'F'], 1: [1, -1, 'C']}, 
       'F': {0: [1, 1, 'D'], 1: [1, 1, 'A']}
    }

on = []

iter = 0
current_index = 0
current_state = 'A'
while iter < 12919244:
    if current_index in on:
        value = 1
    else:
        value = 0
    
    new_value, direction, new_state = ins[current_state][value]
    if value == 0 and new_value == 1:
        on.append(current_index)
    elif value == 1 and new_value == 0:
        on.pop(on.index(current_index))
    
    current_state = new_state
    current_index += direction
    iter += 1

print(len(on))
```
