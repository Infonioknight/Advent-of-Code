```python
conditions, printing = open('day5.txt', 'r').read().split('\n\n')

dependencies = {}

conditions = conditions.split('\n')

for line in conditions:
    x, y = line.split('|')
    if y in dependencies: 
        dependencies[y].append(x)
    else:
        dependencies[y] = [x]

printing = printing.split('\n')
solution = 0

def reorder(pages, dependencies):
    final_pages = []
    i = 0
    while i < len(pages):
        if pages[i] in dependencies:
            unmet = []
            for dep in dependencies[pages[i]]:
                if dep in pages and dep not in final_pages:
                    unmet.append(dep)
            if len(unmet) == 0:
                final_pages.append(pages[i])
                pages.remove(pages[i])
            else:
                for item in unmet:
                    if item in dependencies:
                        buffer = 0
                        for dep in dependencies[item]:
                            if dep in pages and dep not in final_pages:
                                buffer = 1
                        if buffer == 0:
                            final_pages.append(item)
                            pages.remove(item)

                    else:
                        final_pages.append(item)
                        pages.remove(item)
        
        else:
            final_pages.append(pages[i])
            pages.remove(pages[i])
        
    return final_pages


for line in printing:
    printed_pages = []
    pages = line.split(',')
    for page in pages:
        if page in dependencies:
            buffer = 0
            for dep in dependencies[page]:
                if dep in pages and dep not in printed_pages:
                    buffer = 1
                    break
            if buffer == 0:
                printed_pages.append(page)
            else:
                break
        else:
            printed_pages.append(page)

    # if len(printed_pages) == len(pages):
    #     solution += int(printed_pages[len(printed_pages) // 2])

    if len(printed_pages) != len(pages):
        printed_pages = reorder(pages, dependencies)
        solution += int(printed_pages[len(printed_pages) // 2])
print(solution)

       
```

```python
data = open('day7.txt', 'r').read().split('\n')
value = []
combination = []

for item in data:
    first, second = item.split(': ')
    value.append(int(first))
    combination.append([int(item) for item in second.split(' ')])


def joiner(first, second):
    return int(str(first) + str(second))

# I basically want a function, that has (value, start_value, list_of_values)
def recursive_calculator(value, start_value, numbers):
    # Value is the final needed value
    # Start value is the value before the next calculation
    # Numbers are the REVERSED list of numbers to add to our calculation
    if len(numbers) == 0:
        if start_value == value:
            return True
    
    else:
        new_numbers = numbers[:-1]
        if start_value == 0:
            return recursive_calculator(value, numbers[-1], new_numbers)
        
        else:
            if recursive_calculator(value, start_value + numbers[-1], new_numbers) or recursive_calculator(value, start_value * numbers[-1], new_numbers) or recursive_calculator(value, joiner(start_value, numbers[-1]), new_numbers):
                return True
            else:
                return False
        

solution = 0
for i in range(len(value)):
    if recursive_calculator(value[i], 0, combination[i][::-1]):
        solution += value[i]

print(solution)
```

```python
# data = '2333133121414131402'
data = open('day9.txt', 'r').read()
mapping = []

for index, character in enumerate(data):
    if index % 2 == 0: # Condition for a file
        mapping.append([index // 2, int(character), 0])
    else:
        if int(character) > 0:
            mapping.append(['.'] * int(character))

# This basically designs our file system, where each index of the mapping, represents the corresponding character in the data file.
# Each is a container, containing the number of elements of that particular type.

# Now we iterate in reverse (over all the files)
while True:
    data_index = len(mapping) - 1

    while True:
        if (len(mapping[data_index]) == 3 and mapping[data_index][2] == 0) or data_index < 0:
            break
        else:
            data_index -= 1

    
    if data_index < 0:
        break

    # Now, we should have the data index (our next file to work on)
    length_of_file = mapping[data_index][1]
    # print(length_of_file, mapping[data_index])

    buffer = 0
    for space_index in range(0, data_index): # Iterating over all the blocks of space to find smth to fit our data in
        # print(space_index, mapping[space_index])
        if mapping[space_index][0] == '.':
            if len(mapping[space_index]) < length_of_file: 
                continue

            # If same size, replace space with the data and remove the data from old location
            elif len(mapping[space_index]) == length_of_file:
                mapping[data_index][2] = 1 # Has been operated on
                mapping[space_index] = mapping[data_index]
                mapping[data_index] = ['.'] * length_of_file
                buffer = 1
                break
            
            # If space bigger than file, fill as much as needed, keep rest of the space
            elif len(mapping[space_index]) > length_of_file:
                mapping[data_index][2] = 1
                temp = mapping[data_index]
                mapping[data_index] = ['.'] * length_of_file
                mapping.insert(space_index, temp)
                mapping[space_index + 1] = ['.'] * (len(mapping[space_index + 1]) - length_of_file)
                buffer = 1
                break
    
    if buffer == 0:
        mapping[data_index][2] = 1

consolidated_mapping = []
for item in mapping:
    if len(item) == 3 and item[0] != '.':
        value, counts, _ = item
        for _ in range(counts):
            consolidated_mapping.append(value)
    
    else:
        consolidated_mapping += item

solution = 0
for index, item in enumerate(consolidated_mapping):
    if item != '.':
        solution += item * index
print(solution)


    
```

```python
import numpy as np
from collections import deque

data = np.array([list(line) for line in open('inputs/day12test.txt', 'r').read().split('\n')])
# Each entry is PLOT_TYPE: (AREA, PERIMETER)

# Part - 1
price = 0 
seen = set()
for i in range(len(data)):
    for j in range(len(data[0])):
        if (i, j) in seen:
            continue
        else:
            area, perimeter = 0, 0
            queue = deque([(i, j)])
            seen.add((i, j))
            while queue:
                x, y = queue.popleft()
                for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]: # First we calculate the number of fences we need for the current square
                    if x+dx not in range(len(data)) or y+dy not in range(len(data[0])) or data[x+dx, y+dy] != data[x, y]:
                        perimeter += 1

                    if x+dx in range(len(data)) and y+dy in range(len(data[0])) and data[x+dx, y+dy] == data[x, y] and (x+dx, y+dy) not in seen:
                        queue.append((x+dx, y+dy))
                        seen.add((x+dx, y+dy))
                    
                area += 1
            # print(area, perimeter, data[i,j], i, j) # This should be a unique character every time
            price += area * perimeter
print(price)
```

```python
# Part 2 - number of sides of the plot results in cost, not direct fence length
# Fun thing, for a polygon, its interesting to note that n(vertices) = n(edges)
direction_move = {
    (0, 1): [(-1, 0), (1, 0)], # right
    (-1, 0): [(0, -1), (0, 1)], # up
    (1, 0): [(0, 1), (0, -1)], # down
    (0, -1): [(1, 0), (-1, 0)] # left
}

price = 0 
seen = set()
for i in range(len(data)):
    for j in range(len(data[0])):
        if (i, j) in seen:
            continue
        else:
            area = 0
            current_field_polygon = []

            queue = deque([(i, j)])
            seen.add((i, j))
            current_field_polygon.append((i, j))

            while queue:
                x, y = queue.popleft()
                for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]: 
                #     # if x+dx not in range(len(data)) or y+dy not in range(len(data[0])) or data[x+dx, y+dy] != data[x, y]:
                #     #     perimeter += 1

                    if x+dx in range(len(data)) and y+dy in range(len(data[0])) and data[x+dx, y+dy] == data[x, y] and (x+dx, y+dy) not in seen:
                        queue.append((x+dx, y+dy))
                        seen.add((x+dx, y+dy))
                        current_field_polygon.append((x+dx, y+dy))
                    
                area += 1
            
            # Okay, so based off what ive seen, we can iterate over points, we run left absolutely every single time we can, if we cant, we walk straight and if neither, we turn right.
            print(current_field_polygon)
            cx, cy = current_field_polygon[0]
            seen_current = set()
            vertices = 0
            dx, dy = 0, 1

            while (cx, cy, (dx, dy)) not in seen_current: # Assume we start by facing right ->
                # left turn check
                seen_current.add((cx, cy, (dx, dy)))
                if (cx+direction_move[(dx, dy)][0][0], cy+direction_move[(dx, dy)][0][1]) in current_field_polygon:
                    dx, dy = direction_move[(dx, dy)][0]
                    vertices += 1
                
                elif (cx+dx, cy+dy) in current_field_polygon:
                    dx, dy = dx, dy
                elif (cx+direction_move[(dx, dy)][1][0], cy+direction_move[(dx, dy)][1][1]) in current_field_polygon:
                    dx, dy = direction_move[(dx, dy)][1]
                    vertices += 1
                else: 
                    dx, dy = -dx, -dy
                    vertices += 2
                
                cx += dx
                cy += dy
            print(vertices)
            if vertices % 2 == 1:
                vertices -= 1
            price += area * vertices
            # price += area * perimeter
print(price)
```
