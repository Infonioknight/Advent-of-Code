```python
from copy import deepcopy, copy
import sys

sys.setrecursionlimit(20000)

# data = open('Test.txt', 'r').read().split('\n')
data = open("Walkies.txt", 'r').read().split('\n')
start = [0]
end = len(data) - 1
for i in data[0]:
    if i == '.':
        start.append(data[0].index(i))
```

```python
solutions = []

def walkie(r, c, moved):
    if data[r][c] == '#' or (r, c) in moved:
        return
    
    moved.add((r, c))
    # print((r, c), len(moved))
    
    if r == end:
        solutions.append(len(moved))
        print(len(moved) - 1)
        return 

    # if data[r][c] in '><^v':
    #     if data[r][c] == '>' and dc != -1:
    #         walkie(r, c+1, 0, 1, moved)
    #     elif data[r][c] == '<' and dc != 1:
    #         walkie(r, c-1, 0, -1, moved)
    #     elif data[r][c] == '^' and dr != 1:
    #         walkie(r-1, c, -1, 0, moved)
    #     elif data[r][c] == 'v' and dr != -1:
    #         walkie(r+1, c, 1, 0, moved)
    #     else:
    #         return
    
    for dr_move, dc_move in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
        walkie(r+dr_move, c+dc_move, moved.copy())
```
