```python
# day 2 - 2
data = open('d2.txt', 'r').read().split(',')
max_length = 0

id_summer = 0

for item in data:
    lower, upper = item.split('-')
    if max_length < max(len(lower), len(upper)):
        max_length = max(len(lower), len(upper))

# Find all the numbers the LENGTH of the number is divisible by, which means that many ways it can be split into repeating combinations
# Example: Length = 8 -> [1, 2, 4] -> can be split into repeating digits, pairs or quartets
# They also say, the digits must be repeated atleast twice
def factors_of_a_number(number):
    """
    Takes a number input, and gives all the numbers it is divisible by (all not just prime)
    """
    factor_list = []
    for i in range(1, number // 2 + 1):
        if number % i == 0:
            factor_list.append(i)
    return factor_list


length_dictionary = {}
for lengths in range(1, max_length + 1):
    length_dictionary[lengths] = factors_of_a_number(lengths)

# Now for each entry of data, we check that entry's length, map it to all possible repeatable combinations and do a split check
for item in data:
    lower, upper = [int(num) for num in item.split('-')]
    for number in range(lower, upper + 1):
        num_string = str(number)
        current_length = len(num_string)
        if current_length < 2:
            continue
        perfect_splits = length_dictionary[current_length]
        for split in perfect_splits:
            check_list = set([''.join([num_string[chunk:chunk + split]])for chunk in range(0, current_length, split)])
            if len(check_list) == 1:
                id_summer += number
                break

print(id_summer)
```

```python
grid = [list(row) for row in open('d4.txt', 'r').read().split('\n')]
tissue_rolls = []

tossed_counter = 0

# Single pass to identify all the rolls
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == '@':
            tissue_rolls.append((i, j))

while True:
    removed_tissues = []
    for coordinates in tissue_rolls:
        if tissue_crawler(coordinates):
            removed_tissues.append(coordinates)
    
    if len(removed_tissues) == 0:
        break
    else:
        tossed_counter += len(removed_tissues)

    for item in removed_tissues:
        item_index = tissue_rolls.index(item)
        rx, ry = tissue_rolls.pop(item_index)
        grid[rx][ry] = '.'

tossed_counter
```

```python
# day 5
fresh_ranges, test_ingredients = open('d5.txt', 'r').read().split('\n\n')

fresh_ranges = fresh_ranges.split('\n')
test_ingredients = [int(ing) for ing in test_ingredients.split('\n')]

range_map = {}
for num_range in fresh_ranges:
    starting, ending = [int(num) for num in num_range.split('-')]
    if starting in range_map:
        range_map[starting] = max(ending, range_map[starting])
        continue
    range_map[starting] = ending

range_map = {k:v for k, v in sorted(range_map.items())}
iterate_copy = list(range_map.items())

for index, (key1, value1) in enumerate(iterate_copy):
    for (key2, value2) in iterate_copy[index + 1:]:
        if key1 not in range_map.keys():
            continue
        
        if max(key1, key2) <= min(value1, value2):
            range_map[key1] = max(value1, value2)
            del(range_map[key2])
            value1 = max(value1, value2)
            iterate_copy[index] = (key1, max(value1, value2))

fresh = 0
for ingredient in test_ingredients:
    for key, value in range_map.items():
        if ingredient >= key and ingredient <= value:
            fresh += 1
            break

actually_fresh = sum([value - key + 1 for key, value in range_map.items()])   
actually_fresh
```

```python
# Day 6 - 1
from functools import reduce

data = open('d6.txt', 'r').read().split('\n')
# From what I observe, the data has numbers that are of a maximum length of 4

cleaned_data = []

for line in data:
    line = filter(lambda x: x != '', line.split(' '))
    cleaned_data.append(list(line))

total = 0
for iterator in range(0, len(cleaned_data[0])):
    current_symbol = cleaned_data[-1][iterator]
    if current_symbol == '*':
        sol = reduce(lambda x, y: x * y, [int(cleaned_data[i][iterator]) for i in range(0, len(cleaned_data) - 1)])
    else:
        sol = sum([int(cleaned_data[x][iterator]) for x in range(0, len(cleaned_data) - 1)])
    
    total += sol
total
```

```python
column_max_counts = []
for column in range(0, len(cleaned_data[0])):
    current_column = [cleaned_data[row][column] for row in range(0, len(cleaned_data) - 1)]
    column_max_counts.append(len(max(current_column, key=len)))

clean_split_data = []
operators = list(filter(lambda x: x != '', data[-1].split(' ')))

for index, line in enumerate(data[:-1]):
    current_line = []
    line_iterator = 0
    for chunk in column_max_counts:
        current_line.append(line[line_iterator:line_iterator + chunk].replace(' ', '0'))
        line_iterator += chunk + 1
    clean_split_data.append(current_line)

solution = 0
for column in reversed(range(len(operators))):
    # Isolate each column, then go columnwise in that column?
    current_problem = []
    for sub_column in reversed(range(column_max_counts[column])):
        hopefully_correct_value = ''.join([clean_split_data[val][column][sub_column] for val in range(len(clean_split_data))])
        hopefully_correct_value = int(hopefully_correct_value.strip('0'))
        current_problem.append(hopefully_correct_value)

    if operators[column] == "*":
        sub_sol = reduce(lambda x, y: x * y, current_problem)
    else:
        sub_sol = sum(current_problem)
    
    # print(sub_sol)
    solution += sub_sol

solution
```

```python
# day - 7
# Note, merged tachyon beams are counted as a single beam and not multiple instances
manifold = open('d7.txt', 'r').read().split('\n')
active_beams_in_current_row = set([manifold[0].index('S')])
split_counter = 0
# The above starts us off with our source beam.
for line_index in range(1, len(manifold)):
    for index, point in enumerate(manifold[line_index]):
        if point == '^' and index in active_beams_in_current_row:
            active_beams_in_current_row.remove(index)
            active_beams_in_current_row.add(index + 1)
            active_beams_in_current_row.add(index - 1)
            split_counter += 1

print(split_counter)
```

```python
from collections import deque

# Now I think i'd iterate over the beams this time instead of the spaces of the manifold.
# Here, I take each beam, I propagate it in the x (row)(downwards) direction - If I encounter a splitter, I change the beam into two: (x + 1, y - 1)
# and (x + 1, y + 1)
# To keep track of each beam, I'd push it into a deque. So each indiced path will be a separate beam by itself.
# The propagation of the given beam ends when it reaches the last square of the manifold, in which case I shall update a counter for the path.
# Approach takes too long, explodes exponentially, so doesnt work.

timelines = deque([(0, manifold[0].index('S'))])
number_of_alternate_timelines = 0
while timelines: # As long as there is a timeline to be explored
    ct_x, ct_y = timelines.popleft() # Where ct is current timeline
    if ct_x == len(manifold) - 1:
        number_of_alternate_timelines += 1
        continue
    
    ct_x += 1
    if manifold[ct_x][ct_y] == '^': # Splitter
        timelines.append((ct_x, ct_y - 1))
        timelines.append((ct_x, ct_y + 1)) # Both possible timelines are incorporated into the queue
    
    else:
        timelines.append((ct_x, ct_y)) # If unchanged/unmoved

print(number_of_alternate_timelines)
```

```python
# day - 8
# 3D Coordinates. Connect the different boxes together (make N connections as in the question)
# Then we need to run a number of connected components run over the entire graph, to find out the top K groups.
# Multiply the groups together and we get our answer I suppose
from collections import deque
from functools import reduce

data = open('d8test.txt', 'r').read().split('\n')
coordinates = []

# So now I have all the coordinates set
for line in data:
    coordinates.append(tuple([int(num) for num in line.split(',')]))

# Now to get all the distances with the respective nodes.
distances = []
for first in range(len(coordinates) - 1):
    for second in range(first + 1, len(coordinates)):
        x1, y1, z1 = coordinates[first]
        x2, y2, z2 = coordinates[second]
        
        distances.append((coordinates[first], coordinates[second], ((x1 - x2) ** 2) + ((y1 - y2) ** 2) + ((z1 - z2) ** 2)))

distances.sort(key=lambda x: x[2])

# Now to start making my connections and create the graph
graph = {}
number_of_wires = 10
for coord in coordinates:
    graph[coord] = []

for connections in range(number_of_wires):
    source, destination, _ = distances[connections]
    graph[source].append(destination)
    graph[destination].append(source)

# Now to find number of connected components in each sub-graph
seen = set()
connection_record = []

for source in graph.keys():
    if source in seen:
        continue

    connections = 0
    queue = deque([source])
    while queue:
        current = queue.popleft()
        if current in seen:
            continue

        seen.add(current)
        connections += 1
        for destination in graph[current]:
            queue.append(destination)
    
    connection_record.append(connections)

print(connection_record)
print(reduce(lambda x, y: x * y, sorted(connection_record, reverse=True)[:3]))
```
