```python
data = [line.split(' ') for line in open('day22.txt', 'r').read().split('\n')]
cleaned = []

for line in data:
    current = ''
    for information in line:
        if information != '':
            current += information + ' '
    cleaned.append(current.strip())

cleaned = [line.split(' ') for line in cleaned]
```

```python
viable_pairs = 0

for source_index in range (0, len(cleaned)):
    # Source node
    used = int(cleaned[source_index][2][:-1])
    if used != 0:
        for destination_index in range(0, len(cleaned)):
            available = int(cleaned[destination_index][3][:-1])
            if used <= available and destination_index != source_index:
                viable_pairs += 1
print(viable_pairs)
```
