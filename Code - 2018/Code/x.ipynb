```python
data = open('points.txt', 'r').read().split('\n')
position = []
velocity = []

for line in data:
    x = line.replace('<', ':').replace('> ', ':').replace('>', ':')
    x = x.split(':')
    position.append([int(number) for number in x[1].split(',')])
    velocity.append([int(number) for number in x[-2].split(',')])

print(len(position), len(velocity), len(data))
num_points = len(position)
```

```python
import matplotlib.pyplot as plt

condition = True
counter = 0
check = 0

while condition:
    for index in range(len(position)):
        x, y = position[index]
        dx, dy = velocity[index]

        position[index] = [x+dx, y+dy]
    
    points_group = 0
    counter += 1
    for checker in position:
        close = 0
        x, y = checker
        for tx, ty in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            if [x+tx, y+ty] in position:
                close += 1
        
        if close >= 1:
            points_group += 1

    if points_group > check:
        print(points_group)
        check = points_group

    if points_group > 350:
        print(counter)
        condition = False
print(position)
```

```python
grid = []
for i in range(400):
    in_grid = ''
    for j in range(400):
        if [i, j] in position:
            in_grid += '#'
        else:
            in_grid += '.'
    grid.append(in_grid)

for line in grid:
    print(line)
```

```python
class Person(object):
    def __init__(self, name):
        self.name = name
    def say(self, stuff):
        return self.name + ' says: ' + stuff
    def __str__(self):
        return self.name

class Lecturer(Person):
    def lecture(self, stuff):
        return 'I believe that ' + Person.say(self, stuff)

class Professor(Lecturer):
    def say(self, stuff):
        return self.name + ' says: ' + self.lecture(stuff)
         
class ArrogantProfessor(Person):
    def lecture(self, stuff):
        return 'It is obvious that ' + super().say(stuff)
    def say(self, stuff):
        return self.name + ' says: ' + self.lecture(stuff)
    

ae = ArrogantProfessor('eric')
```
